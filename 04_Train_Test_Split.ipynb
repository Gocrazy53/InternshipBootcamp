```python
#import library
import pandas as pd
```

```python
#read data
diabetics = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
#display columns
diabetics.columns
```

```python
#define target y -->> Dependent variable
y = diabetics['diabetes']
```

```python
#define features x --->>Independent variable
x = diabetics.drop(['diabetes'],axis=1)
x
```

```python
# We can also write 'x' as like this
x = diabetics[['pregnancies','glucose','diastolic','triceps','insulin','bmi','dpf','age']]
```

```python
#import train test split function
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2529)
```
