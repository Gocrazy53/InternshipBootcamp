```python
#First Step - Import library
import pandas as pd

#Second Step - Import Data
Salary = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv')
Salary.head()


```

```python
Salary.info()
```

```python
Salary.describe()
```

```python
Salary.columns
```

```python
Salary.shape
```

```python
target_unique_value = Salary['Experience Years'].unique()
print(target_unique_value)
```

```python
target_unique = Salary['Salary'].unique()
print(target_unique)
```

```python
#Define Target variables x and y
x = Salary[['Experience Years']]
y = Salary['Salary']

x.shape
```

```python
y.shape
```

```python
# Step 4 - train test split
# Using sklearn library

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python
x_train
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#Step-5:Select Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Step-6:train model(fit model) -->This is the step where machine learning begins
# This function gives a slope and intercept that results in making the predicting model using the equation ---->>> y(salary) = intercept + m*Experience Years + error
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
#slope(m)
```

```python
#Step-7:prediction ---->>> Here we find the predicted value
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
x_test
```

```python
# step-8:checking accuracy of the model
# How to check -->> from | y_actual - y_pred | = Absoluteerror --->> meansquare of all absolute error gives accuracy of the model
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

So We have 6% of error ... i.e our model accuracy is 94%
